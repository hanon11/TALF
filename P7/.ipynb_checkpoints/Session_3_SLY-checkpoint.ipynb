## Ambiguous Grammars & Debug Mode

Consider an grammar as follow:

```
    expr -> expr PLUS expr
         -> expr MINUS expr
         -> expr TIMUS expr
         -> expr DIVIDE expr
         -> LPAREN expr RPAREN
         -> NUMBER
```

In this case, how is the grouping of the operations? "(3 * 4)+5" or "3 * (4+5)". If it is an ambiguous grammar, you will get messages about "shift/reduce conflict" or "reduce/reduce" conflicts.

A shift/reduce conflict is caused when the parser generator canâ€™t decide whether or not to reduce a rule or shift a symbol on the parsing stack.

````python
class CalcLexer(Lexer):
    # Set of token names.   This is always required
    tokens = { NUMBER}

    # String containing ignored characters
    ignore_newline = r'\n+'
    ignore = ' \t'

    # Regular expression rules for tokens
    literals = {'+', '*', '-', '/'}
    NUMBER = r'\d+'
    
class CalcParser(Parser):
    tokens = CalcLexer.tokens
    debugfile = 'parser.log' 
    @_('E "+" E')
    def E(self, p):
        return p.E0 + p.E1
    
    @_('E "-" E')
    def E(self, p):
        return p.E0 - p.E1
    
    @_('E "/" E')
    def E(self, p):
        return p.E0 / p.E1
    
    @_('E "*" E')
    def E(self, p):
        return p.E0 * p.E1
    
    @_('NUMBER')
    def E(self, p):
        return p.NUMBER
````

In this case, when the parser reaches step 6, it has two options. One is to reduce the rule expr : expr * expr on the stack. The other option is to shift the token + on the stack. Both options are perfectly legal from the rules of the 
context-free-grammar.

* By default, all shift/reduce conflicts are resolved in favor of shifting.
* To resolve ambiguity, especially in expression grammars, SLY allows individual 
tokens to be assigned a precedence level and associativity.
* This is done by adding a variable precedence to the parser class

````python
class CalcParser(Parser):
    tokens = CalcLexer.tokens
    debugfile = 'parser.log' 
    precedence = (
        ('left', "+", "-"),
        ('left', "*", "/")
    )
    @_('E "+" E')
    def E(self, p):
        return p.E0 + p.E1
    
    @_('E "-" E')
    def E(self, p):
        return p.E0 - p.E1
    
    @_('E "/" E')
    def E(self, p):
        return p.E0 / p.E1
    
    @_('E "*" E')
    def E(self, p):
        return p.E0 * p.E1
    
    @_('NUMBER')
    def E(self, p):
        return p.NUMBER
````

Note that shift/reduce conflicts have disappear.

## Syntax Errors

If we care about handling error we have to write grammar rules that include the _error_ token. Consider the next rule:
 
```python
@_('PRINT expr SEMI')
def statement(self, p):
    ...
```
To account for the possibility of a bad expression, you might write an additional grammar rule like this:
```python
@_('PRINT error SEMI')
def statement(self, p):
    print("Syntax error in print statement. Bad expression")

```

In this case, the error token will match any sequence of tokens that might appear up to the first semicolon that is encountered.

* Once the semicolon is reached, the rule will be invoked and the error token will go away.

* This type of recovery is sometimes known as parser resynchronization.

* The error token acts as a wildcard for any bad input text and the token immediately following error acts as a synchronization token.

* It is important to note that the error token usually does not appear as the last token on the right in an error rule.

## Panic Mode ðŸ†˜

It is an alternative scheme to recover information from parser. For example, if we are working with an interpreted language if there is an error, we can try to keep running the script. This scheme is defined under **error(self, p)** method.

There are two methods that can be used:

* __self.errok()__: This resets the parser state so it doesnâ€™t think itâ€™s in error-recovery mode. This will prevent an error token from being generated and will reset the internal error counters so that the next syntax error will call error() again.

```python
def error(self, p):
    if p:
        print("Syntax error at token", p.type)
        # Just discard the token and tell the parser it's okay.
        self.errok()
    else:
        print("Syntax error at EOF")
```

* __self.restart()__: This discards the entire parsing stack and resets the parser to its initial state.

```python
def error(self, p):
    print("Whoa. You are seriously hosed.")
    if not p:
        print("End of File!")
        return

    # Read ahead looking for a closing '}'
    while True:
        tok = next(self.tokens, None)
        if not tok or tok.type == 'RBRACE':
            break
    self.restart()
```
*error* method can also return to parser the next lookahead token.

## AST Constructions

Sly provides no special functions for constructiong and abstract syntax tree, but I do. Using graphviz library we can show a graph of our grammar. There are some TO-DO tasks but it will provide some useful insights about the design of the grammar.

````python
class CalcLexer(Lexer):
    # Set of token names.   This is always required
    tokens = {NUMBER}

    # String containing ignored characters
    ignore_newline = r'\n+'
    ignore = ' \t'

    # Regular expression rules for tokens
    literals = {'+', '*', '-', '/', '=', "(", ")", ";"}
    
    @_(r'\d+')
    def NUMBER(self, t):
        t.value = int(t.value)
        return t
````

````python
class CalcParser(Parser):
    def __init__(self):
        self.vars = {"fact": 0, "suma":0, "E": 0, "+":1, "-":1, "*":1, "/":1}
        self.graph = graphviz.Digraph('Graph', engine='dot', graph_attr={'ordering': 'out'}, node_attr={'style':'filled', 'fillcolor':'#FAFAAA'})

    def addEdge(self, nodes_in, nodes_out, is_terminal=False):
        if is_terminal==False:
            for node_in, node_out in zip(nodes_in, nodes_out):
                self.graph.edge(node_in+'_'+str(self.vars[node_in]), node_out+'_'+str(self.vars[node_out]-1))
                if node_out in ['+','-','/','*']:
                    self.vars[node_out] += 1
            self.vars[node_in]+=1
        else:
            self.graph.edge(nodes_in+'_'+str(self.vars[nodes_in]), nodes_out)
            self.vars[nodes_in]+=1
        
    # Get the token list from the lexer (required)
    tokens = CalcLexer.tokens
    # Grammar rules and actions
    debugfile = 'parser.log' 
    
    @_('E "+" suma')
    def E(self, p):
        self.addEdge(['E','E','E'],['E', '+', 'suma'])
        return p.E + p.suma
    
    @_('E "-" suma')
    def E(self, p):
        self.addEdge(['E','E','E'],['E', '-', 'suma'])
        return p.E - p.suma
    
    @_('suma')
    def E(self, p):
        self.graph.edge('E_'+str(self.vars['E']), 'suma_'+str(self.vars['suma']-1) )
        self.vars['E'] += 1
        return p.suma
    
    @_('suma "/" fact')
    def suma(self, p):
        self.addEdge(['suma','suma','suma'],['suma', '/', 'fact'])
        return p.suma / p.fact
    
    @_('suma "*" fact')
    def suma(self, p):
        self.addEdge(['suma','suma','suma'],['suma', '*', 'fact'])
        return p.suma * p.fact
    
    @_('fact')
    def suma(self, p):
        self.addEdge(['suma'],['fact'])
        return p.fact
    
    @_('NUMBER')
    def fact(self, p):
        self.addEdge('fact',str(p.NUMBER), is_terminal=True)
        return p.NUMBER
````

````python
lexer = CalcLexer()
parser = CalcParser()
text = "3/4+1*2"
result = parser.parse(lexer.tokenize(text))
result
````
