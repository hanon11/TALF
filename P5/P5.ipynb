```python
from sly import Lexer
import pandas as pd
```

```python
class Example(Lexer):
    # errores
    def error(self, t):
        print(t)
        # print(f'Illegal character {t.value[0]}') 
        self.index += 1
    
    tokens = {ID, IF, THEN, ELSE, WHILE, INT, FLOAT, CHAR, FOR, RETURN, COMMENT1, COMMENT2, BARRA, 
              NUMEROS_FLOAT, NUMEROS_ENTEROS, ASTERISCO, SALTO}
    # String containing ignored characters (between tokens)
    ignore = r' \t\[\]\(\)\{\}+-<>=\'\";' # Base ID rule
    ID = r'[a-zA-Z_][a-zA-Z0-9_]*'
    COMMENT1 = r'//[ a-zA-Z_][ a-zA-Z0-9_ \s]*'
    COMMENT2 = r'/\*[ a-zA-Z_][ a-zA-Z0-9_ \s]* \*/'
    BARRA = r'/'
    ASTERISCO = r'\*'
    NUMEROS_FLOAT = r'\d+\.\d+'
    NUMEROS_ENTEROS = r'[0-9]'
    SALTO = r'\n'
    # Special cases
    ID['if'] = IF 
    ID['then'] = THEN
    ID['else'] = ELSE 
    ID['while'] = WHILE
    ID['int'] = INT
    ID['float'] = FLOAT
    ID['char'] = CHAR
    ID['for'] = FOR
    ID['return'] = RETURN
    
tokens = { 'IF': 0, 'ELSE': 0, 'CHAR': 0, 'THEN': 0, 'WHILE': 0, 'INT': 0, 'FLOAT': 0, 'FOR': 0, 'RETURN': 0, 
         'COMMENT1': 0, 'COMMENT2': 0, 'NUMEROS_ENTEROS':0, 'NUMEROS_FLOAT':0}
    
f = open("input.txt")
lineas = f.readlines()
lexer = Example()
i = 1;
for linea in lineas:
    #print(linea)
    for token in lexer.tokenize(linea):
        if token.type != 'ID' and token.type != 'SALTO' and token.type != 'ASTERISCO' and token.type != 'BARRA':
            print(f'Linea: {i}; Type: {token.type}; Value: {token.value}')
            valor_anterior =  tokens[token.type]
            tokens = { **tokens, token.type: valor_anterior+1}
    i = i+1;
f.close()

pd.DataFrame(tokens.items())


```
